Next, we read in the url: 

```python
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

The following reads the html as a list: 

```python
my_data=pd.read_html(url)
    
```

```python
print(my_data[:10])
```

For some weird reason, the who table that we want is the first entry in the list. We convert this first entry to a dataframe:

```python
df=pd.DataFrame(my_data[0])
```

```python
df.head()
```

Next we remove any rows for which the Borough is not assigned:


```python
df=df[df.Borough!="Not assigned"]
df.head()
```

Finally, we replace any unassigned neighbourhoods with the name of the corresponing borough (though I think this never happens in this data): 

```python
df['Neighborhood'] = np.where((df.Neighborhood == 'Not assigned'),df.Borough,df.Neighborhood)
df.head()
```
